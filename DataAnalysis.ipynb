```python
## Comments from code review :
# You could also encode or map the categorical columns values to numerical representations
# and plot a correlation matrix for each subset to see if there were any continuous columns
# that had a high correlation.

df = pd.read_csv("loan_payments.csv",index_col=0)

## Corelated columns
plot = Plotter()

## Corelations on continuous columns
continuous_df = df[['loan_amount', 'int_rate', 'annual_inc', 'funded_amount']]
plot.identify_Collinearity_with_parm(continuous_df, "Correlation heatmap of loans dataframe for continuous columns")

## Corelations on categorical columns
encoded_df = pd.get_dummies(df[['grade', 'purpose', 'home_ownership', 'loan_status']], drop_first=True)
plot.identify_Collinearity_with_parm(encoded_df, "Correlation heatmap of loans dataframe for categorical columns")

# Combine encoded categorical and continuous columns
combined_df = pd.concat([encoded_df, continuous_df], axis=1)
plot.identify_Collinearity_with_parm(combined_df, "Correlation heatmap of loans dataframe for continuous & categorical columns combined")
```
